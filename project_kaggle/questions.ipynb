```python

df = pd.read_csv("data/df_unemployment_rates.csv")
df.rename(columns={'import pandas as pddate': 'date'}, inplace=True)
df['date'] = pd.to_datetime(df['date'])

covid_df = df[(df['date'].dt.year >= 2020) & (df['date'].dt.year <= 2021)]
 #plot
plt.figure(figsize=(12, 8))
plt.plot(covid_df['date'], covid_df['overall_rate'], marker='o') 
plt.title("Monthly Average U.S. Unemployment Rate During COVID-19 (2020–2021)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)  
plt.tight_layout()  
plt.show() 
```
